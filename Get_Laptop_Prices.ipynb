```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://www.officedepot.com/a/browse/dell-laptop-computers/N=5+1462020&cbxRefine=301832&viewAllSkus=true&inStock=true/"
data=requests.get(url).text
soup=bs(data, "html.parser")


tags=soup.find_all("div", class_="sku_description")

def get_laptop_name(tag):
    if len(tag.get_text().strip().split(","))>1:
        return tag.get_text().strip().split(",")[0]
    else:
        return tag.get_text().strip().split("-")[0]

def get_laptop_description(tag):
    if len(tag.get_text().strip().split(","))>1:
        return tag.get_text().strip().split(",")[1:-1]
    else: 
        return tag.get_text().strip().split("-")[1:-1]

def get_price(soup):
    return soup.find("div", class_="od-graphql-price").get_text(" ",strip=True)
    
def get_laptop_price(url):
    laptop_price={}
    data=requests.get(url).text
    soup=bs(data, "html.parser")
      
    try:
        laptop_price["price"]=get_price(soup).split(" ")[0]
    except Exception as e:   
        print(e)
    
    return laptop_price

laptop_info_list=[]
for index, tag in enumerate(tags):
    laptop_info={}
         
    laptop_info["Name"]=get_laptop_name(tag)
    laptop_info["Description"]=get_laptop_description(tag)
    
    url="https://www.officedepot.com"
    relative_url=tag.find("a")["href"]
    full_url=url+relative_url
  
    laptop_price=get_laptop_price(full_url)
    for k, v in laptop_price.items():
        laptop_info[k]=v
        
    laptop_info_list.append(laptop_info)
```
